# Data Extraction in ETL – Assignment Solutions

## Question 1: Types of Data Sources in ETL

- **Structured Data Sources**: Databases like MySQL, PostgreSQL.
- **Semi-structured Data Sources**: JSON, XML files.
- **Unstructured Data Sources**: Text files, images, videos.
- **External Sources**: APIs, web scraping.

**Example**: Sales data from SQL database, customer data from CRM API.

## Question 2: What is Data Extraction?

Data extraction is the process of collecting data from different sources for processing.

**Role in ETL**:
- First step in ETL
- Ensures data availability
- Feeds transformation and loading stages

## Question 3: CSV vs Excel

- **CSV**: Simple, lightweight, faster to process, no formatting.
- **Excel**: Supports formatting, multiple sheets, but heavier.

**ETL Usage**:
- CSV preferred for large-scale processing
- Excel used for reporting and manual analysis

## Question 4: Steps to Extract Data from Relational Database

1. Connect to database
2. Authenticate user
3. Write SQL query
4. Execute query
5. Fetch data
6. Store in staging area

## Question 5: Challenges in Data Extraction

- Data inconsistency
- Large data volume
- Connectivity issues
- Data quality problems

## Question 6: APIs and Real-time Extraction

APIs allow applications to communicate.

**Role in ETL**:
- Fetch real-time data
- Enable automation
- Provide structured data (JSON/XML)

**Example**: Weather API, stock market API

## Question 7: Why Databases Preferred?

- High performance
- Data integrity
- Security
- Scalability
- Easy querying using SQL

## Question 8: Steps for Large CSV Extraction (1GB+)

- Use chunk processing
- Optimize memory usage
- Use efficient libraries (pandas, dask)
- Clean data during extraction
- Parallel processing if possible
